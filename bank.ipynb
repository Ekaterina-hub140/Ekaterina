The program "Bank" does the following: it asks the user: "what would you like to do?" and suggests options: 1)take a mortgage 2)deposit money on a bank account. Ask for the sum of money, to get the input from a user, uses a GUI element in python from 100 to 1000000 to make it look beautiful :) Then we put a second GUI element that lets the user choose the time of the mortgage or money deposit (depends on the user choice), it should give options from 3 months to 5 years with the steps: 3 months, 6 months, 9 months, 1 year, 2 years, 3 years, 5 years. Count the percent and print it for the user ("your mortgage percent is ....") or ("your deposit percent is ....") depending on the choice made by user before(mortgage or deposit). For the mortgage count percent: for sum less then 1000: if the time chosen by the user is 3 months, then the percent is 18%. for each step in time add plus 1 percent. for the sum from 1000 to 10000: the percent is 19% for 3 months and add 1 percent for each step in time. for the sum from 10000 to 1000000 the percent is 20 for 3 months and add 1 percent for each step in time. For the deposit choice: same percent count but 7 percent less for each option.

```python
import tkinter as tk
from tkinter import messagebox, simpledialog, ttk

def calculate_percent(amount, time_period, is_mortgage):
    if amount < 1000:
        base_percent = 18
    elif amount < 10000:
        base_percent = 19
    else:
        base_percent = 20
    
    # Adjust base percent based on time period
    if time_period == '3 months':
        percent = base_percent
    elif time_period == '6 months':
        percent = base_percent + 1
    elif time_period == '9 months':
        percent = base_percent + 2
    elif time_period == '1 year':
        percent = base_percent + 3
    elif time_period == '2 years':
        percent = base_percent + 5
    elif time_period == '3 years':
        percent = base_percent + 7
    elif time_period == '5 years':
        percent = base_percent + 10

    # Adjust percent for deposits
    if not is_mortgage:
        percent -= 7
    
    return percent

def on_submit():
    choice = choice_var.get()
    amount = amount_var.get()
    time_period = time_var.get()
    
    if amount < 100 or amount > 1000000:
        messagebox.showwarning("Input Error", "Amount should be between 100 and 1,000,000.")
        return
    
    percent = calculate_percent(amount, time_period, choice == "Mortgage")
    if choice == "Mortgage":
        result_message = f"Your mortgage percent is: {percent}%"
    else:
        result_message = f"Your deposit percent is: {percent}%"
    
    messagebox.showinfo("Result", result_message)

# Create the main window
root = tk.Tk()
root.title("Finance Calculator")

# Variables
choice_var = tk.StringVar(value="Mortgage")
amount_var = tk.IntVar(value=100)
time_var = tk.StringVar(value="3 months")

# Create widgets
ttk.Label(root, text="What would you like to do?").grid(column=0, row=0, columnspan=2, padx=10, pady=10)

ttk.Radiobutton(root, text="Take a mortgage", variable=choice_var, value="Mortgage").grid(column=0, row=1, padx=10)
ttk.Radiobutton(root, text="Deposit money", variable=choice_var, value="Deposit").grid(column=1, row=1, padx=10)

ttk.Label(root, text="Enter the amount of money (100 - 1,000,000):").grid(column=0, row=2, padx=10, pady=5, columnspan=2)
amount_entry = ttk.Entry(root, textvariable=amount_var)
amount_entry.grid(column=0, row=3, columnspan=2, padx=10)

ttk.Label(root, text="Select the time period:").grid(column=0, row=4, padx=10, pady=5, columnspan=2)
time_options = ["3 months", "6 months", "9 months", "1 year", "2 years", "3 years", "5 years"]
time_combobox = ttk.Combobox(root, textvariable=time_var, values=time_options)
time_combobox.grid(column=0, row=5, columnspan=2, padx=10)

submit_button = ttk.Button(root, text="Submit", command=on_submit)
submit_button.grid(column=0, row=6, columnspan=2, padx=10, pady=10)

# Run the application
root.mainloop()
```
