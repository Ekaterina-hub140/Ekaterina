# BlackJack Game

Description of the game:
2 players (computer+human player)
You can place bets, take more cards or decide to stay. The player who is closer to 21 wins if 21 was not exceeded. Ace is equal to 1 or 11. Jack, Queen and King are equal to 10.


```python
class Accounts():
   
    def __init__(self,owner,value):
        self.owner = owner
        self.value = value
   
    def bet(self,bet):
        self.bet = bet
        
        if self.bet <= self.value:
            self.value = self.value - bet
            print (f"{self.owner} account is now {self.value}")
        else:
            print ("Not enough money on the account")
    
    def win(self,amount):
        self.amount = amount
        
        self.value = self.value + amount
        print (f"{self.owner} account is now: {self.value}")
        
            
```

```python
def blackjack(nums):
    sum = 0
    ace=10
    for x in nums:
        sum = sum + x
    if sum <=21:
        return sum
    else:
        if 11 in nums and sum>21:
            sum = sum - ace
            if sum <=21:
                return sum
            else:
                return "BUST"
                            
        else:
            return "BUST"
    
        
        
```

```python
next_card = 0
def cards():
    #6,7,8,9,10,10,10,10
    global next_card
    
    from random import randint
    next_card = randint(6,11)
    card_type = randint(1,4)
    card_color = randint(1,4)
    
    

    #program names of card color
    if card_color == 1:
        card_color1 = "Spades"
    elif card_color ==2:
        card_color1 = "Diamons"
    elif card_color == 3:
        card_color1 = "Hearts"
    else:
        card_color1 = "Clubs"
        
    if next_card == 10:
    #program Jack, King, Queen, 10
        if card_type == 1:
            card_type1 = "10"
        elif card_type ==2:
            card_type1 = "Jack"
        elif card_type == 3:
            card_type1 = "Queen"
        else:
            card_type1 = "King"
    elif next_card == 11:
        card_type1 = "Ace"
    elif next_card == 9:
        card_type1 = "9"
    elif next_card == 8:
        card_type1 = "8"
    elif next_card == 7:
        card_type1 = "7"
    elif next_card == 6:
        card_type1 = "6"
    
        
      
    
    print (f"{card_type1} of {card_color1} = {next_card} points")

    #use next_card to add value to player's deck of cards
    
```

```python
def play_again():
    result = " "
    while result != 'y' or result !='n':
        result = input("Do you want to continue the game? y or n: ")
        return result == 'y'
    
    
# if True = continue game
```
